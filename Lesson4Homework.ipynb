## 2、 打乱数据集
同时打乱 x,y，保证数据和标签的对应关系\
x = ['快递太慢了！','不好吃','特别难吃','要齁死我了','很划算','下次还来','味道很不错！','香']\
y = ['差评','差评','差评','差评','好评','好评','好评','好评']

```python
import random

x = ['快递太慢了！','不好吃','特别难吃','要齁死我了','很划算','下次还来','味道很不错！','香']
y = ['差评','差评','差评','差评','好评','好评','好评','好评']

length = len(x)

def shuffle(x,y):
    x2 = list()
    y2 = list()
    for i in range(length):
        r = random.randint(0,len(x)-1)
        x2.append(x.pop(r))
        y2.append(y.pop(r))
    
    return x2, y2
        
x, y = shuffle(x,y)
for i,j in zip(x, y):
    print(i, ':', j)
```
